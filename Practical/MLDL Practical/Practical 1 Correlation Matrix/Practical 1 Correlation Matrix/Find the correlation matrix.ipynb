# Practical 1: Find the correlation matrix
The correlation coefficient (ρ) is a measure that determines the degree to which the movement of two different variables is associated. The most common correlation coefficient, generated by the Pearson product-moment correlation, is used to measure the linear relationship between two variables.
Correlation coefficients are indicators of the strength of the linear relationship between two different variables, x and y. A linear correlation coefficient that is greater than zero indicates a positive relationship. A value that is less than zero signifies a negative relationship. Finally, a value of zero indicates no relationship between the two variables x and y.

```python
def corr_coeff(cor_df,x,y):
    # we need to calculate the mean of the 2 columns
    x_bar = cor_df[x].mean()
    y_bar = cor_df[y].mean()
    
    # we need to calculate the sample-deviation of the 2 columns
    x_sd = cor_df[x].std()
    y_sd = cor_df[y].std()
    
    # we need to standardize the columns and multiply and eventually,sum all the products
    total_prod = (((cor_df[x] - x_bar)/x_sd)*((cor_df[y] - y_bar)/y_sd)).sum()
    corr = total_prod/(cor_df.shape[0]-1)
    
    return corr
```

```python
corr_coeff(cor_df,'Sales','Temperature')
```
